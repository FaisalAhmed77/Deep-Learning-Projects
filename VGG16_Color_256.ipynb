# Applying CNN

```python
trb = pd.read_csv("train_1.csv")
trb
```

```python
vb = pd.read_csv("valid.csv")
tb = pd.read_csv("test.csv")
tb
```

```python
yy = list( np.concatenate((trb["diagnosis"], vb["diagnosis"], tb["diagnosis"]), axis = 0))
len(yy)
```

```python
for i in range (len(yy)):
    if yy[i] == 2:
        yy[i] = 1
    if yy[i] == 3:
        yy[i] = 1
    if yy[i] == 4:
        yy[i] = 1
y = np.array(yy)
y
```

```python
# #Spliting the data in 80:20 training to testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 0)
```

```python
y_train
```

```python
import keras
y_train = keras.utils.np_utils.to_categorical(y_train,2)
y_test = keras.utils.np_utils.to_categorical(y_test,2)
```

```python
y_train.shape
```

```python
# Models 
import tensorflow as tf
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Flatten, Dense, Dropout, Conv2D, MaxPooling2D

from tensorflow.keras.applications import VGG16
model_cnn = tf.keras.models.Sequential([
    VGG16(input_shape=(256,256,3), include_top=False),
])
for layer in model_cnn.layers:
  layer.trainable = False

model_cnn.add(Conv2D(64, (3,3), activation='relu'))
model_cnn.add(MaxPooling2D(2,2))
model_cnn.add(Flatten())
model_cnn.add(Dense(64, activation='relu'))
model_cnn.add(Dropout(0.4))
model_cnn.add(Dense(2, activation='sigmoid'))

model_cnn.summary()
```

```python
from tensorflow.keras.utils import plot_model

#model = Model(inputs=[mlp.input, model_cnn.input], outputs=x)

plot_model(model_cnn, "multi_input_and_output_model.png", show_shapes=True)
```

```python

#model = Model(inputs=[mlp.input, model_cnn.input], outputs=x)

loss = keras.losses.BinaryCrossentropy()
model_cnn.compile(loss= loss, optimizer= 'adam', metrics=['accuracy'])
# train the model
print("[INFO] training model...")

h = model_cnn.fit(x=x_train, y=y_train, epochs=50, batch_size=32)
```

```python
score = model_cnn.evaluate(x_test, y_test, verbose=1)
print('loss=', score[0])
print('accuracy=', score[1])
```
